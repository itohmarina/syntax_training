```python
import sqlite3
import pandas as pd
```

```python
df = pd.read_csv('penguins.csv')
with sqlite3.connect('penguins.db') as conn:
  df.to_sql('palmer', con=conn)
```

pandas経由で、sqlite3を使ってDBを作成する。
sqlite3.connect(DB_path)でデータに接続している。DBが存在しなければ新しいDBが作成される。csvをテーブルにするにはpandasのDataFrame.to_sql(table_name, connection)を使用する。
